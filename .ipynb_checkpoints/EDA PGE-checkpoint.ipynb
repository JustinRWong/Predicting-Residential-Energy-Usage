## Exploring Elec

```python
pge_monthly_elec_fn = 'cleaned_data/pge-monthly-elec-by-zip_2013-2020.csv'

pge_monthly_elec = pd.read_csv(pge_monthly_elec_fn).drop('Unnamed: 0', axis=1)
```

```python
display(pge_monthly_elec.head())
pge_monthly_elec.describe()
```

### Monthly KWH Electricity Distribution from 2013-2020

```python
sns.barplot(data=pge_monthly_elec, x='MONTH', y='AVERAGEKWH')
plt.title('PG&E Monthly Average Electricity per Zip Code from 2013-2020')
```

```python
sns.barplot(data=pge_monthly_elec, x='MONTH', y='TOTALKWH')
plt.title('PG&E Monthly Total Electricity per Zip Code from 2013-2020')
```

### KWH Electricity Over time from 2013-2020

```python
# sns.lineplot(data=pge_monthly_elec, x='DATE', y='AVERAGEKWH')
# pge_monthly_elec.plot(x='DATE', y='AVERAGEKWH')
average_elec_across_all_zips = pge_monthly_elec.groupby('DATE').agg('mean').reset_index()
average_elec_across_all_zips.head()
```

```python
average_elec_across_all_zips.plot(x='DATE', y='AVERAGEKWH')
plt.title('PG&E Residential Average Electricity across all Zip Codes from 2013 to 2020')
plt.ylabel('AVERAGE KWH')
```

```python
average_elec_across_all_zips.plot(x='DATE', y='TOTALKWH')
plt.title('PG&E Residential Average Electricity across all Zip Codes from 2013 to 2020')
plt.ylabel('AVERAGE KWH')
# sns.lineplot(data=average_across_all_zips, x='DATE', y='AVERAGEKWH')
```

## Exploring Gas

```python
pge_monthly_gas_fn = 'cleaned_data/pge-monthly-gas-by-zip_2013-2020.csv'

pge_monthly_gas = pd.read_csv(pge_monthly_gas_fn).drop('Unnamed: 0', axis=1)
```

```python
display(pge_monthly_gas.head())
pge_monthly_gas.describe()
```

### Monthly Therms Gas Distribution from 2013-2020

```python
sns.barplot(data=pge_monthly_gas, x='MONTH', y='AVERAGETHERMS')
plt.title('PG&E Monthly Average Gas Usage per Zip Code from 2013-2020')
```

```python
sns.barplot(data=pge_monthly_gas, x='MONTH', y='TOTALTHERMS')
plt.title('PG&E Monthly Total Gas per Zip Code from 2013-2020')
```

### Monthly Therms Gas Usage Overtime from 2013-2020

```python
# sns.lineplot(data=pge_monthly_elec, x='DATE', y='AVERAGEKWH')
# pge_monthly_elec.plot(x='DATE', y='AVERAGEKWH')
average_gas_across_all_zips = pge_monthly_gas.groupby('DATE').agg('mean').reset_index()
average_gas_across_all_zips.head()
```

```python
average_gas_across_all_zips.plot(x='DATE', y='AVERAGETHERMS')
plt.title('PG&E Residential Average Gas across all Zip Codes from 2013 to 2020')
plt.ylabel('AVERAGE THERMS')
```

```python
average_gas_across_all_zips.plot(x='DATE', y='TOTALTHERMS')
plt.title('PG&E Residential Total Gas across all Zip Codes from 2013 to 2020')
plt.ylabel('AVERAGE THERMS')
```
