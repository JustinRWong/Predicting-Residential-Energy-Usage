```python
from sklearn.model_selection import train_test_split
from scipy.optimize import leastsq

from sklearn.linear_model import LinearRegression, Ridge, Lasso

from sklearn.metrics import mean_squared_error
```

## Modeling Average Electricity Across all Zip Codes

```python
pge_monthly_elec_copy = average_elec_across_all_zips.copy()
pge_monthly_elec_copy.columns
```

```python
features = ['MONTH', 'YEAR']
X = pge_monthly_elec_copy[features]
Y = pge_monthly_elec_copy['AVERAGEKWH']

X_train, X_test, Y_train, Y_test = train_test_split(X, Y, test_size=.1, random_state=2021)
```

### Simple Linear Regression

$$ \hat{\vec{y}} = \hat{\vec\beta}*\vec{x}$$

Coefficients in $ \hat{\vec\beta}$  are simply scalars that are fit to the best-fit line. That is, simply solving least squares

$$ \hat{\vec\beta} = (X^TX)^{-1}X^TY$$


```python
slm = LinearRegression()
slm.fit(X_train, Y_train)
slm_train_mse = mean_squared_error(Y_train, slm.predict(X_train))
slm_test_mse = mean_squared_error(Y_test, slm.predict(X_test))

print("Linear Model Train Error: ", slm_train_mse)
print("Linear Model Test Error: ", slm_test_mse)
```

### Sine Linear Regression - Optimized with Ordinary Least Squares (OLS)

reference: https://stackoverflow.com/questions/16716302/how-do-i-fit-a-sine-curve-to-my-data-with-pylab-and-numpy

$$ \hat{y}_{0} = \sigma_{0}sin(x + \omega_{0}) + \bar{x}$$

$$ \hat{y}_{optimized} = \sigma_{optimized}sin(x + \omega_{optimized}) + \bar{x}$$

```python
t = pge_monthly_elec_copy['YEAR'] + (pge_monthly_elec_copy['MONTH']/12)
data = pge_monthly_elec_copy['AVERAGEKWH'] #+ np.random.randn(N) # create artificial data with noise

guess_mean = np.mean(data) ## x_bar: offset 
guess_std = 3*np.std(data)/(2**0.5)/(2**0.5) ## sigma: eyeballed std
guess_phase = -1092.16
guess_freq = 12.52 ## omega: bc 12 months per cycle(year) 
guess_amp = 86.89 ## eyeballed height of lowest to highest 

## Generalize a function
y_hat_initial = guess_std*np.sin(t+guess_phase) + guess_mean


# Define the function to optimize, in this case, we want to minimize the difference
# between the actual data and our "guessed" parameters
optimize_func = lambda x: x[0]*np.sin(x[1]*t+x[2]) + x[3] - data
est_amp, est_freq, est_phase, est_mean = leastsq(optimize_func, [guess_amp, guess_freq, guess_phase, guess_mean])[0]

# recreate the fitted curve using the optimized parameters
data_fit = est_amp*np.sin(est_freq*t+est_phase) + est_mean

fine_t = np.arange(2013,max(t),1/12)
data_fit=est_amp*np.sin(est_freq*fine_t+est_phase)+est_mean
```

### Nonlinear Regression

$$ \hat{y}_{0} = g_{0}\sigma_{0}sin(x + \omega_{0}) + \bar{x}$$

$$ \hat{y}_{optimized} = g_{optimized}\sigma_{optimized}sin(x + \omega_{optimized}) + \bar{x}$$

```python
t = pge_monthly_elec_copy['YEAR'] + (pge_monthly_elec_copy['MONTH']/12)
data = pge_monthly_elec_copy['AVERAGEKWH'] #+ np.random.randn(N) # create artificial data with noise

guess_mean = np.mean(data) ## x_bar: offset 
guess_std = 3*np.std(data)/(2**0.5)/(2**0.5) ## sigma: eyeballed std
guess_phase = 553.13
guess_freq = 925.90 ## omega: bc 12 months per cycle(year) 
guess_amp = 86.79 ## eyeballed height of lowest to highest 
guess_growth = 13.59

## Generalize a function
y_hat_initial_nonlinear = (guess_growth*np.sqrt(guess_std))*np.sin(t+guess_phase) + guess_mean


# Define the function to optimize, in this case, we want to minimize the difference
# between the actual data and our "guessed" parameters
optimize_nonlin_func = lambda x: (x[0] * np.sqrt(x[1]))*np.sin(x[2]*t+x[3]) + x[4] - data
nonlin_est_growth, nonlin_est_amp, nonlin_est_freq, nonlin_est_phase, nonlin_est_mean = leastsq(optimize_func, [guess_growth, guess_amp, guess_freq, guess_phase, guess_mean])[0]

# recreate the fitted curve using the optimized parameters
data_fit_nonlin = (nonlin_est_growth*np.sqrt(nonlin_est_amp))*np.sin(nonlin_est_freq*t+nonlin_est_phase) + nonlin_est_mean

fine_t = np.arange(2013,max(t),1/12)
data_fit_nonlin = (nonlin_est_growth*np.sqrt(nonlin_est_amp))*np.sin(nonlin_est_freq*fine_t+nonlin_est_phase)+nonlin_est_mean
```

### Optimized Ridge Regression

$$ \hat{\vec\beta}^{ridge} = (X^TX + {\lambda}I)^{-1}X^TY$$

```python
x_train_prime, x_val, y_train_prime, y_val = train_test_split(X_train[['MONTH']], Y_train, test_size=0.10, random_state=2021)
```

```python
## your code here
alphas = 10 ** np.linspace(5, 20, num=50, endpoint=True)

def model_ridge(x_train, y_train, x_valid, y_valid, alpha):
    ## initialize Ridge model using corresponding alpha and fit intercept
    m = Ridge(alpha=alpha, fit_intercept=True)
    
    ## Fit to the training data
    m.fit(x_train, y_train)
    
    ## predict on the validation data and calculate MSE
    return mean_squared_error(y_valid, m.predict(x_valid)) ## requires sklearn mean_squared_error


alpha_best = alphas[np.argmin([ model_ridge(x_train_prime, y_train_prime, x_val, y_val, a) for a in alphas]) ]
print("Best Alpha: ", alpha_best)

optimized_ridge = Ridge(alpha=alpha_best, fit_intercept=True)
optimized_ridge.fit(X_train, Y_train)


orm_train_mse = mean_squared_error(Y_train, optimized_ridge.predict(X_train))
orm_validation_mse = mean_squared_error(y_val, optimized_ridge.predict(x_val))
orm_test_mse = mean_squared_error(Y_test, optimized_ridge.predict(X_test))
# X_train, X_test, Y_train, Y_test
print("Optimized Ridge Model Train Error: ", orm_train_mse)
print("Optimized Ridge Model Validation Error: ", orm_validation_mse)
print("Optimized Ridge Test Error: ", orm_test_mse)
```

### Optimized Lasso Regression

```python
x_train_prime, x_val, y_train_prime, y_val = train_test_split(X_train, Y_train, test_size=0.10, random_state=2021)
```

```python
## your code here
alphas = 10 ** np.linspace(5, 20, num=50, endpoint=True)

def model_lasso(x_train, y_train, x_valid, y_valid, alpha):
    ## initialize Ridge model using corresponding alpha and fit intercept
    m = Lasso(alpha=alpha, fit_intercept=True)
    
    ## Fit to the training data
    m.fit(x_train, y_train)
    
    ## predict on the validation data and calculate MSE
    return mean_squared_error(y_valid, m.predict(x_valid)) ## requires sklearn mean_squared_error


lasso_alpha_best = alphas[np.argmin([ model_lasso(x_train_prime, y_train_prime, x_val, y_val, a) for a in alphas]) ]
print("Best Alpha: ", alpha_best)

optimized_lasso = Lasso(alpha=lasso_alpha_best, fit_intercept=True)
optimized_lasso.fit(X_train, Y_train)


o_lasso_m_train_mse = mean_squared_error(Y_train, optimized_lasso.predict(X_train))
o_lasso_m_validation_mse = mean_squared_error(y_val, optimized_lasso.predict(x_val))
o_lasso_m_test_mse = mean_squared_error(Y_test, optimized_lasso.predict(X_test))
# X_train, X_test, Y_train, Y_test
print("Optimized Lasso Model Train Error: ", o_lasso_m_train_mse)
print("Optimized Lasso Model Validation Error: ", o_lasso_m_validation_mse)
print("Optimized Lasso Test Error: ", o_lasso_m_test_mse)
```
