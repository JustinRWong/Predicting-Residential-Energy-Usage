```python
## concatenates the path and fn
def full_path(path, fn):
    return path + "/" + fn

def open_csv_and_merge_to_df(df, fn, log=False):
    if log:
        print("Attempting to open: ", fn)
        
    ## Ensure opening a file with the .csv extension
    assert fn[-4:] == '.csv'
    fn_df = pd.read_csv(fn)
    
    fn_df_rows_count = fn_df.shape[0]
    print("\t\t   {s} rows found".format(s=fn_df_rows_count))
    
    ## convert all columns to upper case to combine
    ## this ensures that 'ZIPCODE' and 'ZipCode' are recognized as the same column when appended
    fn_df.columns = [c.upper() for c in fn_df.columns] 
    
    return df.append(fn_df).reset_index(drop=True), fn_df_rows_count

def parse_pge_csv(pge_csv_fn):
    ## Assumes pge_csv_fn follows this format: 'PGE_YYYY_Q{1-4}_{Electric or Gas}UsageByZip.csv'
    pge, year, quarter, utility = pge_csv_fn.split('_')
    u = utility.split('UsageByZip.csv')[0] ## Gas or Electric
    s = "{y}_{q}_{u}".format(y=year, q=quarter, u=u)
    print("\t\t   {s} parsed.".format(s=s))
    return s
    
```

```python
base_df = pd.DataFrame() ## Must start base_df from scratch in order to not have duplicate data
quarterly_rows_count = {}

## Open each year and quarter in chronological order and build the dataframe
for year in sorted(os.listdir(pge_path)):
    pge_data_year = full_path(pge_path, year)
    print("Extracting Contents for year: {y}".format(y=year))
    print("\tOpening: {p}".format(p=pge_data_year))
    
    for pge_year_q in sorted(os.listdir(pge_data_year)):
        ## only build the csv files
        if pge_year_q[-4:] == ".csv":
            pge_year_q_fn = full_path(pge_data_year, pge_year_q)
            print("\t\t>{fn}".format(fn=pge_year_q_fn))
            
            base_df, pge_year_q_rows = open_csv_and_merge_to_df(base_df, pge_year_q_fn)
            k = parse_pge_csv(pge_year_q)
            quarterly_rows_count[k] = pge_year_q_rows
            
    print("\n")
```

```python
# pp.pprint(quarterly_rows_count)
## Sanity check to ensure that all rows read in the individual dataframes are in the base_df cumulatively
print("rows in base df: ", base_df.shape)
print("values summed individually: ", sum(quarterly_rows_count.values()))
quarterly_rows_count
```

```python
## Specify which columns have certain data types
convert_dict = {
    'ZIPCODE': int,
    'MONTH': int,
    'YEAR': int,
    'TOTALCUSTOMERS': int,
    'TOTALKWH': int,
    'AVERAGEKWH': int,
    'TOTALTHERMS': int,
    'AVERAGETHERMS': int
    
}
## Drop columns we dont want to keep
bad_index = 'Unnamed: 0'
if bad_index in full.columns:
    full_with_cities = full.drop('Unnamed: 0', axis=1)
    
## Get rid of commas for the columns that are supposed to be integers
full.replace(',', '', regex=True, inplace=True)
full.fillna(0, inplace=True)

numerical_columns = ['TOTALCUSTOMERS', 'TOTALKWH', 'AVERAGEKWH', 'TOTALTHERMS', 'AVERAGETHERMS']
for num_col in numerical_columns:
    full[num_col] = full[num_col].apply(pd.to_numeric, errors='coerce')

full = full.astype(convert_dict)
full = full.astype({'ZIPCODE': str})
full.head()
```
